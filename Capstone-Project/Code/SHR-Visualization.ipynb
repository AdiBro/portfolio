# Visualization Notebook for the Supplementary Homicide Report Dataset

```python
# Necessary imports:
import numpy as np
import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import seaborn as sns
```

```python
# Display plots in Jupyter Notebook
%matplotlib inline

# Increase maximum rows and columns to 999 (instead of the default 60)
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

```python
# Load in the Supplementary Homicide Report (SHR)
shr = pd.read_csv('/Users/adibronshtein/Documents/Datasets/Capstone/SHR1976_2015.csv')
```

```python
shr.drop_duplicates(inplace=True)
```

## Pie Chart of Solved vs. Unsolved Cases

```python
## Rate of crimes solved
solved = pd.DataFrame(shr, columns = ['Solved']) 
resolution = solved.stack().value_counts()
ax = resolution.plot(kind = 'pie', figsize=(8,8), fontsize=15, title = 'Percentage of Homicides Solved',startangle = 10,autopct='%.3f')
ax.set_ylabel('')
```

## Bar Chart of Solved vs. Unsolved Cases

```python
plt.figure(figsize=(12,6))
yearCases = shr.Solved.value_counts()
yearCases.sort_index().plot(kind='bar')
```

## Horizontal Bar Chart of Solved vs. Unsolved Cases


```python
plt.figure(figsize=(12,6))
yearCases = shr.Solved.value_counts()
yearCases.sort_index().plot(kind='barh')
```

## Murder Rates by States:

```python
plt.figure(figsize=(12,6))
sns.countplot(x='State', data=shr)
plt.xticks(rotation=90, fontsize=15)
plt.yticks(fontsize=15)
plt.title('Homicide Rates by State')
plt.xlabel('State')
plt.ylabel('Number of Homicides')
plt.show()
```

## Murder Rates by Years:

```python
plt.figure(figsize=(12,6))
yearMRD = shr['Year'].value_counts()
yearMRD.sort_index().plot(kind='line')
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Year', data=shr)
plt.xticks(rotation=90, fontsize=15)
plt.title('Homicide Rates by Year')
plt.xlabel('Year')
plt.ylabel('Number of Homicides')
plt.show()
```

```python
# Number of cases by year and solved/unsolved
plt.figure(figsize=(16,8))
solved_year_crosstab = pd.crosstab(shr.Year, shr.Solved)
for item in solved_year_crosstab.columns:
    plt.plot(solved_year_crosstab.index, solved_year_crosstab[item])
    
plt.xlabel('Year', fontsize=20)
plt.ylabel('Solved and Unsolved Cases', fontsize=20)
plt.legend(fontsize=15)
```

## Murder Rates by Months:

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Month', data=shr, order=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
plt.xticks(rotation=90)
plt.title('Homicide Rates by Months')
plt.xlabel('Month')
plt.ylabel('Number of Homicides')
plt.show()
```

## Murder Rates by Type of Agency:

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Agentype', data=shr)
plt.xticks(rotation=90)
plt.title('Number of Murder Cases by Type of Agency')
plt.xlabel('Agency type')
plt.ylabel('Number of Homicides')
plt.show()
```

## Murder Rates by Type of Weapos:

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Weapon', data=shr)
plt.xticks(rotation=90, fontsize=15)
plt.title('Type of Weapons Used in Murders')
plt.xlabel('Type of Weapon')
plt.ylabel('Number of Homicides')
plt.show()
```
